```python
import pandas as pd
import seaborn as sb
import matplotlib as plt
import numpy as np
import sklearn
import matplotlib.pyplot as mtplt
import re
```

```python
tweetData = pd.read_csv('featureEngineered.csv', index_col=False)
tweetData
```

```python
tweetData = tweetData.drop(['Unnamed: 0', 'index'], axis = 1)
```

```python
def count_punctuations(tweet):
    punctuations='!"#$%&\'()*+,-./:;<=>?@[\]^_`{|}~'
    d=dict()
    for i in punctuations:
        d[str(i)+' count']=tweet.count(i)
    return d

for i, tweet in enumerate(tweetData['tweet']):
  dictPunct = count_punctuations(tweet)
  for key,val in dictPunct.items():
    tweetData.loc[i,key] = val
```
